Por ejemplo, el mensaje "MARCIANOS" lo codificaramos como "21 8 19 5 4 8 1 6 2"

Para poder enviar mensajes, debemos introducir los dgitos en una matriz 3x3. En este ejemplo, la matriz sera:

$$A=\begin{bmatrix}21 & 8 & 19 \\ 5 & 4 & 8 \\ 1 & 6 & 2\end{bmatrix}$$

A continuacin, hay que multiplicar por la izquierda esta matriz por otra proporcionada por el Gobierno:

$$C=\begin{bmatrix}2 & 0 & 1 \\ 1 & 0 & 1 \\ 0 & 1 & 0\end{bmatrix}$$

Por tanto, el mensaje resultante sera:

$$Z = CA=\begin{bmatrix}2 & 0 & 1 \\ 1 & 0 & 1 \\ 0 & 1 & 0\end{bmatrix}\begin{bmatrix}21 & 8 & 19 \\ 5 & 4 & 8 \\ 1 & 6 & 2\end{bmatrix}=\begin{bmatrix}43 & 22 & 40 \\ 22 & 14 & 21 \\ 5 & 4 & 8\end{bmatrix}$$

El receptor del mensaje deber decodificarlo multiplicando la matriz $Z$ por la matriz $C^{-1}$ (por la izquierda):

$$C^{-1}Z=A$$

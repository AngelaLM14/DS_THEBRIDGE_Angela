#### Utilidades

````python
### Formas de incluir variables dentro de strings

var1 = 4
var2 = 3

print("Quiero", var1, "manzanas y", var2, "peras.", sep=' ')
print("Quiero %s manzanas y %s peras." %(var1, var2))
print('Quiero {} manzanas y {} peras.'.format(var1, var2))
print(f"Quiero {var1} manzanas y {var2} peras.")
````

````python
### Cambiar tipo dentro de una lista
lista_str = ["2","3"]
lista_int = [int(lista_str[0]),int(lista_str[1])]
print(lista_int)

elemento1 = int(lista_str[0])
elemento2 = int(lista_str[1])

lista_int_2 = [elemento1, elemento2]
print(lista_int_2)
````

````python
### Método join. para convertir listas a strings: 

lista_join = ["a", "b", "c", "d"]

print(lista_join)

string_join = (' '.join(lista_join))

print(string_join)

print(type(string_join))
````

#### Utilidades

Las **tuplas son inmutables** pero si accedes a un elemento en una lista dentro de una tupla:



````python
# LISTAS POR REFERENCIA
variable = "Hola"
y = variable
print(y)
y = "Adiós"
print("variable = ",variable)
print("y = ", y)

# En las listas cuando se iguala una variable lista con otra, se hace una referencia, no se copia el contenido de la variable.
ejer_2 = [1,2,3,4,5]
x = ejer_2
x[2] = 12
print("x = ",x)
print("ejer_2 = ", ejer_2)

# Por eso, si queremos hacer una modificación solo en la lista copiada, debemos utilizar el .copy()
lista_copia = ejer_2.copy()
lista_copia[0] = 6
print("ejer_2 = ", ejer_2)
print("lista_copia = ", lista_copia)

# Ocurre lo mismo con los diccionarios
````

## Ejercicio 2
Haz un programa que detecte todos los duplicados de un elemento en una lista:
* Utiliza una variable duplicado para asignarle el valor del elemento del que queremos encontrar todos sus duplicados. 
* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
`````

`````python
ejer_3_min = []
for palabra in ejer_3: 
    palabra = palabra.lower()
    ejer_3_min.append(palabra)
print(ejer_3_min)
`````

`````python
duplicados = []
indices = []
for i, palabra in enumerate(ejer_3_min):
    if ejer_3_min.count(palabra) > 1 and palabra not in duplicados:
        duplicados.append(palabra)
        for j, word in enumerate(ejer_3_min):
            if word == palabra:
                indices.append(j)
print(duplicados,indices)
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
DNI= 55555555
palabra='TRWAGMYFPDXBNJZSQVHLCKE'
print('La letra del DNI es: ', palabra[DNI%23])
`````

`````python
def dni_valido(dni):

    # Verificar la longitud del DNI
    if (7 <= (len(str(dni))-1) <= 8):
        return True

    # Lista de letras del DNI
    letras_dni = 'TRWAGMYFPDXBNJZSQVHLCKE'

    # Extraer el número y la letra del dni
    dni_num = int(dni[:-1])
    dni_letra = dni(-1)

    #Calcular la letra válida para los dígitos de este dni
    letra_valida = letras_dni[dni_num % 23]

    # Verificar si la letra del DNI coincide con la correspondiente por el número
    if letra_valida == dni_letra.upper():
        return true

dni = input("Indique su DNI:")
if dni_valido(dni) == True:
    print("DNI válido")
else:
    print("El DNI indicado no es válido")


`````

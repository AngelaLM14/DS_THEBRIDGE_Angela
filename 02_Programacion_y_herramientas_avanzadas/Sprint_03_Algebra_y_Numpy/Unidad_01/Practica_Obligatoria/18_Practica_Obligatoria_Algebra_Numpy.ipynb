```python
cantidades = np.array([[6,5,3,1],
                      [3,6,2,2],
                      [3,4,3,1]])
precios = np.array([[1.5,1],
                   [2,2.5],
                   [5,4.5],
                   [16,17]])
```

```python
# Lista de personas
lista_personas = ["Rafa", "Nacho", "Clara"]

# Calcular los gastos de cada persona en cada tienda
gastos_manolo = np.dot(cantidades, precios[:, 0])
gastos_starbucks = np.dot(cantidades, precios[:, 1])

# Recomendar a cada persona dnde comprar
for i, persona in enumerate(lista_personas):
    if gastos_manolo[i] < gastos_starbucks[i]:
        print(f"{persona} debe comprar en Manolo Bakes (Gasto: {gastos_manolo[i]})")
    elif gastos_manolo[i] > gastos_starbucks[i]:
        print(f"{persona} debe comprar en Starbucks (Gasto: {gastos_starbucks[i]})")
    else: 
        print(f"{persona} gastara lo mismo en ambas tiendas (Gasto: {gastos_starbucks[i]} y {gastos_manolo[i]})")
```

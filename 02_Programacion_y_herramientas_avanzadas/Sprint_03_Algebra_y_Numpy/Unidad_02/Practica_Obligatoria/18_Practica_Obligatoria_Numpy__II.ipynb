```python
# Crear tablero con funcin
def crear_tablero(filas, columnas):
    tablero = np.full((filas, columnas), ' ', dtype=str)
    return tablero
```

```python
# Posicionar barcos con funcin
def posicionar_barcos(tablero, barcos):
    for barco in barcos:
        for posicion in barco:
            tablero[posicion[0], posicion[1]] = 'O'
```

```python
# Posicionar disparos con funcin 

def recibir_disparo(tablero, coordenada):
    if tablero[coordenada] == 'O':
        tablero[coordenada] = 'X'
    elif tablero[coordenada] == ' ':
        tablero[coordenada] = '-'
    else:
        tablero[coordenada] = tablero
```

```python
# Posicionar otro barco de 4 posiciones dentro del tablero, pero **de manera ALEATORIA.**

def posicionar_barco_aleatorio(tablero, longitud_barco):
    while True:
        orientacion = random.choice(['N', 'S', 'E', 'O'])
        if orientacion in ['N', 'S']:
            fila = random.randint(0, tablero.shape[0] - 1)
            columna = random.randint(0, tablero.shape[1] - 1)
            posiciones = [(fila + i, columna) for i in range(longitud_barco)]
        else:
            fila = random.randint(0, tablero.shape[0] - 1)
            columna = random.randint(0, tablero.shape[1] - longitud_barco)
            posiciones = [(fila, columna + i) for i in range(longitud_barco)]

        # Verificar si las posiciones estn disponibles
        ocupadas = any(tablero[pos[0], pos[1]] == 'O' for pos in posiciones)

        if not ocupadas:
            for pos in posiciones:
                tablero[pos[0], pos[1]] = 'O'
            break
```

```python
###     COLOCA BARCOS, VERSIN MS INTELIGENTE      ###

def coloca_barco_plus(tablero,barco): 
## Nos devuelve el tablerosi puede colocar el barco, si no, devuelve False y avisa por pantalla
    tablero_temp = tablero.copy()
    num_max_filas = tablero.shape[0]
    num_max_columnas = tablero.shape[1]
    for pieza in barco:
        fila = pieza[0]
        columna = pieza [1]
        if fila < 0 or fila >= num_max_filas: 
            print(f"No puedo poner la pieza {pieza} por que se sale del tablero")
            return False
        if columna < 0 or columna >= num_max_columnas:
            print(f"No puedo poner la pieza {pieza} por que se sale del tablero")
            return False
        if tablero[pieza] == "O" or tablero[pieza] == "X":
            print(f"No puedo poner la pieza {pieza} porque hay otro barco")
            return False
        tablero_temp[pieza] = "O"
    return tablero_temp
```

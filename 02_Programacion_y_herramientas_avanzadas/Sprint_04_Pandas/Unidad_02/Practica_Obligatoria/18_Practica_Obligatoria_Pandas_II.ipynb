1. Clasifiquemos los vuelos en Larga Distancia (LD), Media Distancia (MD) y Regionales (R). Con los siguientes lmites: ms de 9000km, entre 1500 y 9000 kms y resto de vuelos. Aade una columna al `DataFrame` que tenga esa clasificacin por vuelo.

```python
def cat_distancia(row):
    distancia = row["Distancia"]
    if distancia > 9000:
        categoria = "LD"
    elif 1500 < distancia <= 9000:
        categoria = "MD"
    else:
        categoria = "R"
    return categoria
```

```python
df_sin_nan["Cat_distancias"] = df_sin_nan[["Distancia"]].apply(cat_distancia, axis = 1)
```

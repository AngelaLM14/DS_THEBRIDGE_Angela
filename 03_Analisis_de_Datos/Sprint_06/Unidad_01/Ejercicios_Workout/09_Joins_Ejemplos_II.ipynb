### Preparacin

Como en la sesin anterior, importamos libreras, creamos la conexin a la base de datos, creamos el cursor y nuestra funcin para lanzar queries. Adems tambin recuperamos el modelo de datos para tener claro que cruces pueden hacerse. 

```python
import pandas as pd
import sqlite3

# Conectamos con la base de datos chinook.db
connection = sqlite3.connect("./data/chinook_joins.db")

# Obtenemos un cursor que utilizaremos para hacer las queries
cursor_bootcamp = connection.cursor()
```

```python
def sql_query(query):
    cursor_bootcamp.execute(query)
    ans = cursor_bootcamp.fetchall()
    names = [description[0] for description in cursor_bootcamp.description]
    return pd.DataFrame(ans,columns=names)
```

### Right Join

Este es el JOIN en el que manda la tabla de la "derecha", la que ponemos detrs del  `JOIN`. Es la tabla de la que queremos mantener todas las filas y pegarle campos de otra tabla aunque no haya cruce, en cuyo caso se rellenarn con nulos (NULL o NaN si lo pasamos a Pandas)

```python
# Hagamos el inverso del anterior pero por la derecha: Queremos pegar a cada disco las canciones
# Manda la tabla de lbumes (albums) 
# sobre el de canciones (tracks)
# Como en la sesin anterior la clave de cruce es AlbumId (Foreing_Key en tracks, Primary_Key en albums)
query = ''' 
SELECT b.*, a.name
FROM tracks AS a
RIGHT JOIN albums AS b
ON a.albumid = b.albumid
'''
df_rj = sql_query(query)
```

```python
df_rj.info()
```

Aqu las puntualizaciones son dos:
1. Aqu no hay albumid que s est en albums y no est en tracks por eso no hay nulos en el campo `name`
2. Como ahora comprobaremos el nmero de regitros ha crecido porque ahora hay tantas lineas por album como canciones tenga el album, mientras que la tabla anterior de albums el campo AlbumId era nico.

```python
query = ''' 
SELECT * 
FROM albums
'''
df_albums = sql_query(query)
df_albums.info()
```

```python
print(df_albums.AlbumId.isunique)
print(df_rj.AlbumId.is_unique)
```

### Inner Join

Veamos ahora un ejemplo de Inner Join, es decir aquel en el que slo quiero quedarme con las filas que tengan valores comunes en las dos tablas (aquellas filas de una tabla cuyo identificador no aparezca en la clave de la otra no se tendrn en cuenta)

```python
# Vamos a quedarnos ahora con las canciones para las cuales si haya disco
query = '''
SELECT tracks.*, album.title
FROM tracks
INNER JOIN albums
ON tracks.AlbumId = albums.AlbumId
'''
df_inner = sql_query(query)
df_inner.info()
```

Aqu lo que podemos observar es:
1. No hay nulos, claro se cogen filas que existen en una y otra tabla.
2. Se ha reducido el nmero de salidas respecto al left join que hicimos (aunque no te des cuenta ahora si te haces una select de tracks podrs comprobar que tiene  3503 cacniones). Ha quitado las canciones que no tienen album.

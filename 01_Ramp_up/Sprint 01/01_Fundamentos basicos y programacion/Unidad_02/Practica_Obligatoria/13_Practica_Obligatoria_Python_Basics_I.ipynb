## PARTE I

En este ejercicio debes simular un chatbot al que le haremos pedidos de pizzas.

1. El chatbot tiene que saludar con un: "Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana Cuntas pizzas  familiares desea?"
2. El ususario tiene que introducir un nmero de pizzas en una variable llamada 'pizz_familiar'
3. El chatbot debe responder: "Estupendo, cuntas pizzas medianas desea?" y guardar el nmero de pizzas medianas en otra variable 'pizz_mediana'
4. Adems debe responder del chatbot: "Estupendo, se estn preparando 'pizz' pizzas familiares y 'pizz_mediana' pizzas medianas. Digame su direccin"
5. El ususario tiene que introducir una direccion en formato String en otra variable llamada 'direcc'
6. El programa debe calcular el monto total 
7. Respuesta final del chatbot: "Le mandaremos las '(nmero total de pizzas)' pizzas a la direccin 'direcc'. Sern <precio total a pagar en euros>. Muchas gracias por su pedido."
         


```python
pizz_familiar = int(input("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestra pizzas es de 8,95 para la familiar o de 9,90 para la mediana Cuntas pizzas  familiares desea?"))

pizz_mediana = int(input("Estupendo, cuntas pizzas medianas desea?"))

# NO SE PUEDEN METER VARIABLES EN UN INPUT

print('Estupendo, se estn preparando ', pizz_familiar, ' pizzas familiares y ' , pizz_mediana , ' pizzas medianas.')

direccion = input('Dgame su direccin.')
          
precio_familiar = 8.95

precio_mediana = 9.90

coste_pedido = pizz_familiar * precio_familiar + pizz_mediana * precio_mediana 

total_pizz = pizz_familiar + pizz_mediana

print("Le mandaremos las " , total_pizz , " pizzas a la direccin" , direccion , ". Sern " , coste_pedido ,  ". Muchas gracias por su pedido.")

```

Crea un programa que realice las siguientes acciones:

1. Solicite al usuario que introduzca 3 nmeros enteros separados por comas (por ejemplo: 1,2,3).
2. Convierta esa entrada en una lista de nmeros enteros.
3. Utilizando funciones built-in, determine el nmero mximo, el nmero mnimo y la suma total de los nmeros introducidos.
4. Determine la verdad o falsedad de la siguiente afirmacin: "El nmero mximo menos el nmero mnimo es igual a la suma total dividida por 5".
5. Muestre los resultados utilizando print.

```python
enteros_str = input("Introduzca tres nmeros enteros separados por una coma")

lista = enteros_str.split(",")

print(lista)
```

```python
lista_int = [int(lista[0]), int(lista[1]), int(lista[2])]

print(lista_int)
```

```python

maximo = max(lista_int)
print(maximo)
minimo = min(lista_int)
print(minimo)

suma=sum(lista_int)
print(suma)

v_f = (maximo - minimo) == ((suma)/5)
print(v_f)
```
